1: Write a regular expression that can find all amounts of money in a text.
Your expression should be able to deal with different formats and
currencies, for example £50,000 and £117.3m as well as 30p, 500m euro,
338bn euros, $15bn and $92.88. Make sure that you can at least detect
amounts in Pounds, Dollars and Euros. (20pts)
For full marks: include the output of a Python program that applies your regular expression to the following BBC News Web site:
https://www.bbc.co.uk/news/business-41779341


```python
# we are now importing the libraries
from bs4 import BeautifulSoup
import requests

url="https://www.bbc.co.uk/news/business-41779341"

# having a ger request to search for the raw HTML content
html_content = requests.get(url).text
# Parse the html content
data = BeautifulSoup(html_content, "lxml")
print(data.prettify())
```

```python
#printing the data which is been converted from html to text format
print(data.text)
```

```python
import nltk
```

```python
#importing regular expression from nltoolkit
from nltk import re
```

```python
prices = re.findall(r'[EUR|Euro|EURO|euro|EUROS|Euros|euros]?[$|£|€]\s?[0-9]*\,?\.?\s?[0-9]*\s?[a-zA-Z]*',data.text )
```
